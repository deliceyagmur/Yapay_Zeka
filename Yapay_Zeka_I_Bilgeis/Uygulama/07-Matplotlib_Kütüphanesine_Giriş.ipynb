```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("ev_satislari.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data_head_100 = data.head(100)

plt.figure(figsize=(15,6))

plt.plot(data_head_100.datesold,data_head_100.price,
         color="black", 
         linewidth=5, 
         linestyle="--", 
         marker="o", 
         markersize=10, 
         markerfacecolor="green",
         markeredgewidth=4,
         markeredgecolor="blue")

plt.plot(data_head_100.datesold,data_head_100.bedrooms,
         color="red", 
         linewidth=5, 
         linestyle="--", 
         marker="o", 
         markersize=10, 
         markerfacecolor="green",
         markeredgewidth=4,
         markeredgecolor="grey")

plt.xticks(rotation=90)

plt.title("Yıllara göre ev fiyatları")
plt.xlabel("Tarih")
plt.ylabel("Fiyat")

plt.show()
```
